#loop implementation

```python
%%time
def three_nearest_to_average(task_list): #todo n nearest

  stop_signs=['',None,' '] # also  ' ' * n, where n - amount of repeats
  emptiness=[i not in stop_signs for i in task_list]

  for i in range(len(task_list)-1):
    prev=task_list[i-1]
    cur=task_list[i]
    next=task_list[i+1]

    if prev==task_list[-1]:
      prev='0'

    li=[prev,cur,next]
    li=[i!='0' for i in li]
    #print(li)
    avg=(float(prev)+cur+next) / sum(li)
    #if i%2==0:
    print(prev,cur,next,round(avg,2) ) #rounding_weakness

  last_one=( task_list[-1]+task_list[-2] )/sum(emptiness[-2::]) #todo try except if not int values
  print(task_list[-1],task_list[-2],last_one)

three_nearest_to_average(task)
# charlie=[60,50]
# andy=[60,80,50]
# alex=[50,40,80]
# sindy=[40,80]
```
